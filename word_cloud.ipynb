```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd

df = pd.read_excel('data/yt_comments_tokenized.xlsx')
df.head()
```

```python
import ast

full_tokens = []

for i in range(len(df)):
    tokens = ast.literal_eval(df.loc[i, 'tokenized'])
    full_tokens += tokens

full_text = ''

for token in full_tokens:
    if token != '사유리':
        full_text = full_text + ' ' + token
```

```python
wc_full = WordCloud(font_path='JejuGothic.ttf', background_color = 'white').generate(full_text)
plt.figure(figsize=(12,12))
plt.imshow(wc_full, interpolation='bilinear')
plt.axis('off')
plt.show()
```
